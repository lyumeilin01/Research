```python
import pretty_midi
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display
import miditok
from miditok import REMI, TokenizerConfig
from miditoolkit import MidiFile, Instrument, Note
#64, 69
midi_path = './result/from_scratch_64.midi'
```

```python
midi_path = '../LoadMIDI/Script/test.mid'
```

```python
# Load an existing MIDI file (or create a new one)
midi_data = pretty_midi.PrettyMIDI(midi_path)

# Create a new percussion instrument track (MIDI program 0 = standard percussion kit)
beat_track = pretty_midi.Instrument(program=0, is_drum=True)

# Get the time signature changes (if any), otherwise assume default 4/4
time_signatures = midi_data.time_signature_changes

# Get the total length of the piece in seconds
total_time = midi_data.get_end_time()

# Find beat times — returns a list of times at each beat (quarter note)
beat_times = midi_data.get_beats()

# Add a note (click) at every bar start
# A bar start is every Nth beat depending on the time signature
# Let's compute bar length in beats
if len(time_signatures) > 0:
    # Take first time signature as reference (complex handling if it changes later)
    ts = time_signatures[0]
    beats_per_bar = ts.numerator
else:
    beats_per_bar = 4  # default to 4/4

# Now add a beat note (MIDI pitch 37 is a side stick — a nice click sound)
for i, beat_time in enumerate(beat_times):
    if i % beats_per_bar == 0:  # every bar start
        note = pretty_midi.Note(
            velocity=100, pitch=37, start=beat_time, end=beat_time + 0.05
        )
        beat_track.notes.append(note)

# Add this beat track to the MIDI file
midi_data.instruments.append(beat_track)

# Save the modified MIDI
midi_data.write('your_file_with_beats1.mid')
```
